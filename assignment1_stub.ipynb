```python
import csv
import dateutil.parser
import json
import math
import matplotlib.pyplot as plt
import numpy
import sklearn
from collections import defaultdict
from sklearn import linear_model
```

```python
path = "used_cars_train.json"
f = open(path)

data_train = []

for l in f:
    d = json.loads(l)
    data_train.append(d)

f.close()
```

```python
data_train[20]
```

```python
def feature(d):
    feat = [1] # Constant feature
    posting_year = dateutil.parser.parse(d['posting_date']).year
    model_year = int(d['year'])
    feat.append(posting_year - model_year)
    return feat
```

```python
feature(data_train[0])
```

```python
X_train = numpy.array([feature(d) for d in data_train])
y_train = numpy.array([d["price"] for d in data_train])

model = sklearn.linear_model.LinearRegression(fit_intercept=False)
model.fit(X_train, y_train)
```

```python
theta = model.coef_
theta
```

```python
path = "used_cars_test.json"
f = open(path)

data_test = []

for l in f:
    d = json.loads(l)
    data_test.append(d)

f.close()
```

```python
X_test = numpy.array([feature(d) for d in data_test])
y_test = numpy.array([float(d["price"]) for d in data_test])

y_pred = model.predict(X_test)
sse = sum([x**2 for x in (y_test - y_pred)])
mse = sse / len(y_test)
print("MSE",int(mse))
```
